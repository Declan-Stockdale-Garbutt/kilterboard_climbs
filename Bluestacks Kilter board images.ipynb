```python
import pyautogui
import time
import pytesseract
from PIL import Image, ImageGrab
import cv2
import os
import pandas as pd
import numpy as np
```

```python
# set path
path = "C:/Users/Declan/Pictures/Kilterboard_problems/"
```

```python
# read in all file names from path
completed_problems = [image[:-4] for image in os.listdir(path)]
inital_completed_problems = completed_problems
```

```python
# click on window in task bar
pyautogui.moveTo(720, 739)
time.sleep(0.2)

pyautogui.click() 
time.sleep(0.2)
pyautogui.moveTo(691, 717)
time.sleep(0.5)


counter = 0
count_placeholder = 1

while counter < 14360:

    # Move mouse to side of image 
    pyautogui.moveTo(892, 391)

    # Get name of file
    pic = ImageGrab.grab(bbox=(471,122,774,150))

    # Run pytesseract
    climb_name = pytesseract.image_to_string(pic).replace('\n\x0c',"").replace('?',"").replace("|",'I').replace("/",' ').replace("|",'I').replace("**",' ').replace("\&",' ').replace("*",' ')

    print(climb_name)
    
    # save file
    screenshot = ImageGrab.grab(bbox=(430,122,818,724))
    
    # some file names have non english names or some weird punctuation, need to save with placeholder for now
    try:
        screenshot.save(f"{path}{climb_name}.png")
    except:
        screenshot.save(f"{path}placeholder_{count_placeholder}.png")
        count_placeholder +=1

    # move to next climb
    pyautogui.moveTo(780, 402)
    time.sleep(0.2)
    pyautogui.click()
    time.sleep(0.3)
    pyautogui.press('left') # swipes right
    
    completed_problems.append(climb_name)
    
    counter += 1
```

```python
pyautogui.moveTo(720, 739)
time.sleep(0.2)

pyautogui.click() 
time.sleep(0.2)
pyautogui.moveTo(691, 717)
time.sleep(0.5)

# move to search bar
#pyautogui.moveTo(599, 150)

count_placeholder = 0
for climb in sorted_list:
    
    
    # move to search bar
    pyautogui.moveTo(599, 150)
    
    # Write climb name in search bar
    pyautogui.write(climb) 
    
    # trickery waiing for type to be complete
    
    wait_type = True
    while wait_type:
        test_value_type = np.asarray(ImageGrab.grab(bbox=(482,135,629,160))).sum() 
        
        if test_value_type == 2739948:
            print('waiting')
            time.sleep(0.1)
            
        else:
            print('break')
            break

    # what happens if no result?
    
    time.sleep(0.5)
    test_climb_exists = np.asarray(ImageGrab.grab(bbox=(438,189,800,208))).sum() 
    time.sleep(0.5)
    
    if  test_climb_exists == 5014062: # no result
        
        # append climb for later storage
        non_existant_climbs.append(climb)
        #print(f"{climb} doesn't exists, continue")
        pass
        
    else:
    
        # click first climb
        pyautogui.moveTo(744, 219)
        pyautogui.click()

        # Move mouse to side of image 
        pyautogui.moveTo(892, 391)
        time.sleep(0.5)

        # save file
        screenshot = ImageGrab.grab(bbox=(430,122,818,724))

        # Save image
        print('climb_name',climb)
        try:
            climb_ = climb.replace("?","").replace(":","").replace("/","").replace('"',"")
            screenshot.save(f"{path}{climb_}.png")
            print('saved as is')
        except:
            screenshot.save(f"{path}placeholder_{count_placeholder}.png")
            print('saved placeholder')
            count_placeholder +=1

        time.sleep(.5)

        # Escape back to search 
        pyautogui.press('escape') 
        time.sleep(3)

    # move to search bar
    pyautogui.moveTo(478, 151)
    time.sleep(0.5)
    pyautogui.click()
    time.sleep(0.5)
    
    
    # Delete name - takes a while, trying trickery
    pyautogui.keyDown('ctrl')  # hold down the shift key
    pyautogui.press('z')
    pyautogui.keyUp('ctrl')
            
    
    time.sleep(0.5)
```
